### `list.insert(index, element)`

The `insert()` method adds an element at a specified position in the list. It takes two arguments:

*   `index`: The position at which to insert the element. Elements from this index onwards are shifted to the right.
*   `element`: The item to be inserted into the list.

```python
my_list = [10, 20, 30, 40]
print(f"Original list: {my_list}")

# Insert 'new_value' at index 2
my_list.insert(2, 'new_value')
print(f"List after inserting 'new_value' at index 2: {my_list}")

# Insert 'first' at the beginning (index 0)
my_list.insert(0, 'first')
print(f"List after inserting 'first' at index 0: {my_list}")

# Insert 'last' at the end (index equal to current length)
my_list.insert(len(my_list), 'last')
print(f"List after inserting 'last' at the end: {my_list}")
```

### `list.extend(iterable)`

The `extend()` method appends all the elements from an iterable (e.g., a list, tuple, set, or string) to the end of the current list. This is different from `append()` which adds the iterable as a single element.

It takes one argument:

*   `iterable`: Any iterable (e.g., list, tuple, string) whose elements will be added to the list.

```python
my_list = [1, 2, 3]
print(f"Original list: {my_list}")

# Extend with another list
another_list = [4, 5, 6]
my_list.extend(another_list)
print(f"List after extending with another_list: {my_list}")

# Extend with a tuple
my_tuple = (7, 8)
my_list.extend(my_tuple)
print(f"List after extending with a tuple: {my_list}")

# Extend with a string (each character is added as an element)
my_string = "ABC"
my_list.extend(my_string)
print(f"List after extending with a string: {my_list}")

# Compare with append()
list_for_append = [10, 20]
list_for_append.append([30, 40])
print(f"List after append() with a list: {list_for_append}")
```

### `list.append(element)`

The `append()` method adds a single `element` to the end of the list. It modifies the list in place and does not return a new list.

It takes one argument:

*   `element`: The item to be added to the end of the list. This can be any data type, including another list or object.

```python
my_list = [10, 20, 30]
print(f"Original list: {my_list}")

# Append a single integer
my_list.append(40)
print(f"List after appending an integer: {my_list}")

# Append a string
my_list.append("hello")
print(f"List after appending a string: {my_list}")

# Append another list (as a single element)
my_list.append([50, 60])
print(f"List after appending another list: {my_list}")

# Append a boolean
my_list.append(True)
print(f"List after appending a boolean: {my_list}")
```

### `list.remove(element)`

The `remove()` method removes the *first* occurrence of a specified `element` from the list. It modifies the list in place and does not return any value. If the specified `element` is not found in the list, it raises a `ValueError`.

It takes one argument:

*   `element`: The item whose first occurrence is to be removed from the list.

```python
my_list = [10, 20, 30, 20, 40, 50]
print(f"Original list: {my_list}")

# Remove an existing element
my_list.remove(30)
print(f"List after removing 30: {my_list}")

# Remove the first occurrence of a duplicate element
my_list.remove(20)
print(f"List after removing the first 20: {my_list}")

# Attempt to remove an element that does not exist
try:
    my_list.remove(99)
except ValueError as e:
    print(f"Error: {e} - element not found in list.")

print(f"Final list: {my_list}")
```

### `list.pop([index])`

The `pop()` method removes and returns the item at the given `index` from the list. If no index is specified, `pop()` removes and returns the last item in the list.

It modifies the list in place.

It takes one optional argument:

*   `index`: The index of the item to be removed. If omitted, the last item is removed. If the index is out of range, an `IndexError` is raised.

```python
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

# Pop the last element (no index specified)
popped_item_last = my_list.pop()
print(f"List after popping last element: {my_list}, Popped item: {popped_item_last}")

# Pop an element at a specific index (e.g., index 1, which is 20)
popped_item_index = my_list.pop(1)
print(f"List after popping element at index 1: {my_list}, Popped item: {popped_item_index}")

# Attempt to pop an element from an invalid index
try:
    my_list.pop(10) # Index 10 is out of range
except IndexError as e:
    print(f"Error: {e} - cannot pop from invalid index.")

print(f"Final list: {my_list}")
```

### `list.count(element)`

The `count()` method returns the number of times a specified `element` appears in the list. It does not modify the list.

It takes one argument:

*   `element`: The item whose occurrences are to be counted in the list.

```python
my_list = [1, 2, 3, 2, 4, 2, 5]
print(f"Original list: {my_list}")

# Count occurrences of an existing element
count_2 = my_list.count(2)
print(f"Number of times 2 appears: {count_2}")

# Count occurrences of a non-existing element
count_9 = my_list.count(9)
print(f"Number of times 9 appears: {count_9}")

# Count occurrences of another existing element
count_3 = my_list.count(3)
print(f"Number of times 3 appears: {count_3}")

# Count an element of a different type (which won't be found)
count_str = my_list.count('2')
print(f"Number of times '2' appears: {count_str}")
```

### `list.sort(key=None, reverse=False)`

The `sort()` method sorts the items of the list in place. It modifies the list directly and does not return any value. By default, it sorts in ascending order.

It takes two optional arguments:

*   `key`: A function to be called on each list element prior to making comparisons. The default value is `None`.
*   `reverse`: A boolean value. If `True`, the list elements are sorted in descending order. The default value is `False`.

```python
my_list = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"Original list: {my_list}")

# Sort in ascending order (default)
my_list.sort()
print(f"List after default sort (ascending): {my_list}")

my_list = [3, 1, 4, 1, 5, 9, 2, 6]
# Sort in descending order
my_list.sort(reverse=True)
print(f"List after sort (descending): {my_list}")

my_list_str = ["banana", "apple", "cherry", "date"]
print(f"Original string list: {my_list_str}")

# Sort strings by length
my_list_str.sort(key=len)
print(f"List after sort by length: {my_list_str}")

my_list_tuples = [('apple', 3), ('banana', 1), ('cherry', 2)]
print(f"Original list of tuples: {my_list_tuples}")

# Sort list of tuples by the second element
my_list_tuples.sort(key=lambda item: item[1])
print(f"List after sort by second element of tuple: {my_list_tuples}")
```

### `list.sort()` and `sorted()` Time Complexity

Both `list.sort()` (which sorts a list in-place) and `sorted()` (which returns a new sorted list) use an algorithm called **Timsort**.

**Timsort** is a hybrid stable sorting algorithm, derived from merge sort and insertion sort, designed to perform well on many kinds of real-world data.

*   **Worst-case time complexity:** O(n log n)
    *   This is generally the maximum time required to sort a list of `n` elements.
*   **Average-case time complexity:** O(n log n)
    *   For randomly ordered data, Timsort performs very efficiently.
*   **Best-case time complexity:** O(n)
    *   If the list is already sorted or nearly sorted, Timsort can detect this and achieve linear time complexity.
*   **Space complexity:** O(n)
    *   In the worst case, Timsort requires auxiliary space proportional to the number of elements in the list (though it can be O(log n) for certain cases depending on implementation details and data characteristics).

**Key Characteristics of Timsort:**
*   **Stable:** It preserves the relative order of equal elements.
*   **Adaptive:** It takes advantage of existing runs (monotonically increasing or decreasing sequences) within the data, which makes it perform extremely well on partially sorted arrays.

In summary, `list.sort()` and `sorted()` are highly optimized for Python and provide excellent performance for a wide range of input data.

### `sorted(iterable, key=None, reverse=False)`

The `sorted()` function returns a *new sorted list* from the items in any iterable (e.g., list, tuple, string, set, dictionary keys). Unlike the `list.sort()` method, which sorts a list in-place and returns `None`, `sorted()` creates and returns a completely new list, leaving the original iterable unchanged.

It takes the following arguments:

*   `iterable`: Any iterable object (e.g., list, tuple, string, set, dictionary).
*   `key` (optional): A function to be called on each element of the iterable prior to making comparisons. The default value is `None`.
*   `reverse` (optional): A boolean value. If `True`, the list elements are sorted in descending order. The default value is `False`.

#### Key Differences between `sorted()` and `list.sort()`

1.  **Return Value**: `sorted()` returns a *new list*, while `list.sort()` modifies the original list *in-place* and returns `None`.
2.  **Applicability**: `sorted()` can be used on *any iterable*, whereas `list.sort()` can only be used on *lists*.
3.  **Original Data**: `sorted()` preserves the original iterable, creating a new one. `list.sort()` changes the original list.

```python
my_list = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"Original list: {my_list}")

# Using sorted() to get a new sorted list (ascending)
sorted_list_asc = sorted(my_list)
print(f"New list after sorted() (ascending): {sorted_list_asc}")
print(f"Original list remains unchanged: {my_list}")

# Using sorted() to get a new sorted list (descending)
sorted_list_desc = sorted(my_list, reverse=True)
print(f"New list after sorted() (descending): {sorted_list_desc}")

my_tuple = (10, 5, 20, 15)
print(f"\nOriginal tuple: {my_tuple}")
# sorted() works on tuples, returns a list
sorted_tuple_as_list = sorted(my_tuple)
print(f"Sorted tuple (as a new list): {sorted_tuple_as_list}")

my_string = "python"
print(f"\nOriginal string: '{my_string}'")
# sorted() works on strings, returns a list of characters
sorted_string_as_list = sorted(my_string)
print(f"Sorted string (as a new list of chars): {sorted_string_as_list}")

words = ["apple", "banana", "date", "cherry"]
print(f"\nOriginal words list: {words}")
# Sort words by length using the key argument
sorted_by_length = sorted(words, key=len)
print(f"Words sorted by length: {sorted_by_length}")
```

### Using the `key` argument in `sorted()` with a `lambda` function

A `lambda` function is a small, anonymous function defined with the `lambda` keyword. It can take any number of arguments, but can only have one expression. They are often used as arguments to higher-order functions (functions that take other functions as arguments), like `sorted()`, `map()`, `filter()`, etc.

When you use `key=lambda item: expression` with `sorted()`, you are telling Python to apply the `expression` to each `item` in the iterable, and then use the result of that expression as the basis for comparison during the sorting process.

This is incredibly useful for sorting complex data structures like lists of tuples or dictionaries based on a specific element or a calculated value, without modifying the original items themselves.

```python
# Example: Sorting a list of dictionaries

students = [
    {'name': 'Alice', 'age': 25, 'grade': 'A'},
    {'name': 'Bob', 'age': 22, 'grade': 'C'},
    {'name': 'Charlie', 'age': 23, 'grade': 'B'},
    {'name': 'David', 'age': 25, 'grade': 'B'}
]

print(f"Original list of students:\n{students}")

# Sort by 'age' in ascending order
sorted_by_age = sorted(students, key=lambda student: student['age'])
print(f"\nSorted by age:\n{sorted_by_age}")

# Sort by 'grade' in descending order
sorted_by_grade_desc = sorted(students, key=lambda student: student['grade'], reverse=True)
print(f"\nSorted by grade (descending):\n{sorted_by_grade_desc}")

# Sort by 'age' (primary) and then by 'name' (secondary) for students with same age
sorted_by_age_then_name = sorted(students, key=lambda student: (student['age'], student['name']))
print(f"\nSorted by age, then by name:\n{sorted_by_age_then_name}")

# Sort by the length of the 'name'
sorted_by_name_length = sorted(students, key=lambda student: len(student['name']))
print(f"\nSorted by name length:\n{sorted_by_name_length}")
```

### Tuples

A `tuple` is an ordered, immutable collection of items. This means once a tuple is created, its elements cannot be changed, added, or removed. Tuples are defined by enclosing the elements in parentheses `()`, separated by commas.

**Key Characteristics of Tuples:**

*   **Ordered**: The items have a defined order, and that order will not change.
*   **Immutable**: Once created, the elements of a tuple cannot be modified.
*   **Allow Duplicates**: Tuples can contain duplicate items.
*   **Heterogeneous**: Tuples can store items of different data types (integers, strings, lists, etc.).

Tuples are often used for data that should not change, such as coordinates, database records, or returning multiple values from a function.

```python
# 1. Creating Tuples

# Empty tuple
empty_tuple = ()
print(f"Empty tuple: {empty_tuple}, Type: {type(empty_tuple)}")

# Tuple with mixed data types
my_tuple = (1, "hello", 3.14, True, [10, 20])
print(f"My tuple: {my_tuple}, Type: {type(my_tuple)}")

# Tuple with a single item (note the comma!)
single_item_tuple = ("single",)
print(f"Single item tuple: {single_item_tuple}, Type: {type(single_item_tuple)}")

# If you omit the comma, it's just a string in parentheses
not_a_tuple = ("single")
print(f"Not a tuple (just a string): {not_a_tuple}, Type: {type(not_a_tuple)}")

# Tuple creation without parentheses (tuple packing)
tuple_packed = 1, 2, "three"
print(f"Tuple packed: {tuple_packed}, Type: {type(tuple_packed)}")

# 2. Accessing Elements

# Tuples are indexed like lists (zero-based)
print(f"\nFirst element: {my_tuple[0]}")
print(f"Last element: {my_tuple[-1]}")

# Slicing also works
print(f"Slice from index 1 to 3: {my_tuple[1:4]}")

# 3. Immutability (Attempting to modify will raise an error)

# Uncommenting the line below will raise a TypeError
# my_tuple[0] = 5
# print(f"Attempted modification: {my_tuple}")

print("\nTuples are immutable: Attempting to modify elements will raise a TypeError.")

# However, if a tuple contains mutable elements (like a list), those mutable elements can be changed
mutable_element_tuple = (1, [2, 3], 4)
print(f"Original tuple with mutable element: {mutable_element_tuple}")
mutable_element_tuple[1].append(5)
print(f"Tuple after modifying its mutable list element: {mutable_element_tuple}")

# 4. Tuple Unpacking

# Assigning tuple elements to individual variables
a, b, c = tuple_packed
print(f"\nUnpacked: a={a}, b={b}, c={c}")

# Swapping variables easily
x, y = 10, 20
print(f"Before swap: x={x}, y={y}")
x, y = y, x
print(f"After swap: x={x}, y={y}")

# 5. Common Tuple Operations

tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

# Concatenation
concatenated_tuple = tuple1 + tuple2
print(f"\nConcatenated tuple: {concatenated_tuple}")

# Repetition
repeated_tuple = tuple1 * 2
print(f"Repeated tuple: {repeated_tuple}")

# Length
print(f"Length of my_tuple: {len(my_tuple)}")

# Membership test
print(f"'hello' in my_tuple: {'hello' in my_tuple}")
print(f"99 in my_tuple: {99 in my_tuple}")

# 6. Tuple Methods

duplicate_tuple = (1, 2, 2, 3, 4, 2)

# count()
print(f"\nNumber of 2s in duplicate_tuple: {duplicate_tuple.count(2)}")

# index()
print(f"Index of first 3 in duplicate_tuple: {duplicate_tuple.index(3)}")

# Will raise a ValueError if element is not found
try:
    duplicate_tuple.index(99)
except ValueError as e:
    print(f"Error when finding index of non-existent element: {e}")
```
