```python
#find mean of values row and column wise

import numpy as np
matrix = np.random.randint(1, 90, (5, 5))
print(matrix)
print("Mean of values row wise: ", np.mean(matrix, axis=1))
print("Mean of values column wise: ", np.mean(matrix, axis=0))
```

```python
#replace all even numbers with -1
import numpy as np
matrix = np.random.randint(1, 90, (5, 5))
print(matrix)
matrix[matrix % 2 == 0] = -1
print(matrix)
```

```python
#sort the array row amd column wise

import numpy as np
matrix = np.random.randint(1, 90, (5, 5))
print(matrix)
print("Sorted array row wise: ", np.sort(matrix, axis=1))
print("Sorted array column wise: ", np.sort(matrix, axis=0))
```

```python
#find row and column wise sum seperately

import numpy as np
matrix = np.random.randint(1, 90, (5, 5))
print(matrix)
print("Row wise sum: ", np.sum(matrix, axis=1))
print("Column wise sum: ", np.sum(matrix, axis=0))
```

```python
#QUESTION-02

#create a 5*5 array (stock levels)
import numpy as np
stock_levels = np.random.randint(20, 80, (5, 5))
print(stock_levels)
```

```python
# create another 5*5 array (demand)
import numpy as np
demand = np.random.randint(1, 90, (5, 5))
print(demand)
```

```python
#Remaining stock = stock - demand
import numpy as np
remaining_stock = stock_levels - demand
print(remaining_stock)
```

```python
# replace negative values ->0 ( out of stock)

import numpy as np
remaining_stock[remaining_stock < 0] = 0
print(remaining_stock)

modified = np.where(remaining_stock < 0, 0, remaining_stock)
print(modified)
```

```python
#total shortage

import numpy as np
total_shortage = np.sum(np.where(remaining_stock * -1,))
print(total_shortage)
```

```python
#PRODUCT WITH ZERO STOCK

import numpy as np
product_with_zero_stock = np.where(remaining_stock == 0)
print(product_with_zero_stock)

product_shortage = np.sum (np.where(modified == 0,1,0), axis = 0)
print(product_shortage)
```

```python
# sort each row based on remaining stock

import numpy as np
sorted_remaining_stock = np.sort(remaining_stock, axis=1)
print(sorted_remaining_stock)
```
